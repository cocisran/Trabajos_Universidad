## Ejercicio

Haz un gráfico de barras para un número arbitrario $n$ de simulaciones de este problema (considera un número de fósforos fijo igual a $10$). Cada barra debe reprersentar la frecuencia relativa de la cantidad de cerillos restantes que quedan en la caja no vacía. Observése que una vez que alguna de las cajas se ha vaciado, en la otra caja solo puede quedar 0 o bien 1, 2, 3, 4, 5, 6, 7, 8, 9 ó 10 cerillas. Tienes hasta la media noche de hoy.

```python
import random

def banach(n):
    cajaDerecha = cajaIzquierda = n #Creamos dos cajitas con n cerillos
    numeroIntentos = 0
    
    while True: #mientras no hemos intentando sacar cerilos de una caja vacia
        numeroIntentos+=1
        if random.randint(0, 1) == 0:
            if(cajaDerecha == 0): #la caja derecha esta vacia
                break
            cajaDerecha -=1
        else:
            if(cajaIzquierda ==0): #la caja izquierda esta vacia
                break
            cajaIzquierda -=1
            
    if cajaIzquierda == 0:
        return "caja Izquierda","caja derecha",numeroIntentos,cajaDerecha
    else:
        return "caja Derecha","caja izquierda",numeroIntentos,cajaIzquierda
```

```python
def printResultado(r):
    print("Se vacio la ",r[0] , " nos dimos cuenta tras intentar sacar ",r[2], " fosforos")
    print("En la caja ",r[1]," quedaron ",r[3], "fosforos")
```

```python
from collections import defaultdict

numeroFosforos = 100
n = 100000

frecuencias = defaultdict(int)
for x in range(n):
    experimento = banach(numeroFosforos)
    frecuencias[experimento[3]] += 1
frecuencias = sorted(frecuencias.items())

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
resultado = [str(i) for i,j in frecuencias]
frecuencia = [j/n for i,j in frecuencias]
ax.bar(resultado,frecuencia)
plt.title("Frecuencias relativa en "+str(n)+" repeticiones")
plt.show()


```
